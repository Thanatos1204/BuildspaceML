```python
data = data.fillna(method='bfill')
data.head()
```

```python
#to find correlation matrix for better plotting
heat = data.corr(numeric_only=True)
sns.heatmap(heat,cmap="rocket",annot=True)
```

```python
heat
```

```python
#to find the number of houses between each price range.
plt.hist(data['median_house_value'],color='hotpink',ec='blue')
```

```python
figure = plt.figure(figsize=(10,8))

plt.scatter(data['median_income'],data['median_house_value'],color='hotpink')
plt.xlabel('median_income')
plt.ylabel('median_house_value')
plt.title('ScatterPlot:  Median Income  vs. median_house_value')
plt.show()
```

```python
#Line plot for Population analysis. population vs median_income

sns.histplot(x=data['population'],y=data['median_income'], color='blue')
plt.xlabel('Population')
plt.ylabel('Median Income')
plt.title('Histogram: Population vs. Median Income')
plt.show()
```

```python
#Pair Plots to give plotting of each numeric variable with one another.
sns.pairplot(data)
```

```python
x=data['median_house_value']
y=data['total_rooms']

def generate_ap():
    a1 = 0  
    d = 10  
    n = 20640    
    ap = np.array([a1 + (i * d) for i in range(1, n + 1)])

    return ap

colors = generate_ap()

plt.scatter(x,y,c=colors,cmap='viridis')
plt.xlabel('Median_House_Value')
plt.ylabel('Total_Rooms')
plt.title('Scatter Plot: Median House Value vs. Total Rooms')
plt.show()
```
